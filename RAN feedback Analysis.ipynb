```python
clean_string = [word for word in cleaned_text if word not in stop_list]
clean_string
```

```python
# print(*not_stop)
str1 = " "
str1 = str1.join(clean_string)
```

# word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud().generate(str1)
plt.figure(figsize = (12, 12)) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.show()
```
